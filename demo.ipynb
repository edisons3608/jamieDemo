```python
data = pd.read_csv("sample.csv")
#pd.read_csv will import the data as a 2D array.
```

```python

"""Since the "data" variable is a custom Pandas DataFrame object, we use
.iloc[] to index the array"""

#data.iloc[whatever row,0 for x column and 1 for y column]

print(data.iloc[0,0])
#output = 1
print(data.iloc[0,1])
#output = 3.888889
```

```python

#plt.figure will make the plot which you can display ur points on.
#change dpi for resolution.
fig = plt.figure(dpi=500)
#use plt.plot() for continuous lines
#plt.plot(data.iloc[:,0],data.iloc[:,1],marker="o",markersize=0.2,label="Graph")
#use plt.scatter() to display discrete points

#data.iloc[:,0] returns a 1D array of all the x-values of the points
#data.iloc[:,1] returns a 1D array of all teh y-values of the points.

plt.scatter(data.iloc[:,0],data.iloc[:,1],linewidths=0.02,label="Graph")

plt.legend(loc="upper right",prop={'size': 6})
#research plt.title(), plt.xlabel(), and plt.ylabel() to make the graph more deatiled.

#you can also research plt.xticks and plt.yticks to control the frequency of the tick marks.


#call plt.show() to show the figure.
plt.show()
```
